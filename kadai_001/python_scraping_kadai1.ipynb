```python
import requests
from bs4 import BeautifulSoup

url = "https://news.yahoo.co.jp/articles/6164ad60c5f0b26696aa199372eceac2f842e4bb"
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")

meta = soup.select_one('meta[property="og:title"]')
title = meta["content"].strip() if meta else soup.select_one("h1").get_text(strip=True)

print(title)
```
