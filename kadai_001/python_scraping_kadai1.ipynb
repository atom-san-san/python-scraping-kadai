# 新しいセクション

```python
import requests
from bs4 import BeautifulSoup

url = "https://news.yahoo.co.jp/articles/6164ad60c5f0b26696aa199372eceac2f842e4bb"

# ① HTML を取得
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# ② 本文のある領域を指定
body = soup.select_one("article div.article_body")

# ③ 段落ごとに抽出して結合
if body:
    text = "\n".join(p.get_text(strip=True) for p in body.select("p"))
else:
    text = ""

print(text)
```
